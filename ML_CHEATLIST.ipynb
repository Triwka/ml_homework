## Методика класстеризации

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler
!pip install imblearn
from imblearn.over_sampling import SMOTE , ADASYN
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('https://github.com/Triwka/ml_homework/blob/main/data_for_ml_cheatlist/example_oversampling_bank_exit_cleaned.csv?raw=true')
```

pivot

```python
pivot_balance = pd.pivot_table(df, index = ['CustomerId'],  values = 'Balance')
pivot_balance
```

функция создания кластеров с помощью KMeans


```python
def auto_kmeans(cluster_lst,df): ##сюда передается датафрейм, созданный с помощью pivot

  clusters_dict = {}
  for cluster in cluster_lst:

    model = KMeans(n_clusters = cluster, random_state = 42)

    clusters = pd.DataFrame([]) ## создание пустого датафрейма, который будет наполнятся по мере выполнения функции

    #clusters['CustomerId'] = df.columns ## создание первой колонки в пустом датафрейме, в которой значения будут - названия колонок из датафрейма, созданного пивотом

    clusters['n'] = model.fit_predict(df.values) ## предсказание к какому кластеру относился бы объект

    clusters_dict[str(cluster) + '_' + 'cluster' + '_' + 'df'] = clusters.set_index(pivot_balance.index) ## назначение колонки в качестве индексов
    #clusters_dict[str(cluster) + '_' + 'cluster' + '_' + 'df'] = clusters.set_index('CustomerId') нужно образать внимание на индексы, чтобы join сработал
  return clusters_dict
```

```python
df_test = df.copy(deep=True)
```

функция обучения(используем randomforest с помощью gridsearch)

```python
def grid_study(df):
  features = df.drop(['Exited'],axis = 1).reset_index(drop=True) ## features = X, дропаем колонку с целевыми признаками
  target = df.reset_index(drop=True)['Exited'] ## target = y, присваиваем переменной target колокну с целевыми признаками
  X_train, X_test, y_train, y_test = train_test_split(features, target, random_state = 42, test_size=0.25)

  clf = RandomForestClassifier(random_state=42)
  params = {'n_estimators':[100,150,200],'max_depth': [2,3,4], 'criterion':['gini','entropy'],'min_samples_split':[2,3,4]}
  grid_model = GridSearchCV(clf, params, error_score = 'raise')

  grid_model.fit(X_train,y_train)

  y_preds = grid_model.predict(X_test)

  return f1_score(y_test, y_preds)
```

```python
customer_dict=auto_kmeans([2,3,4,5],pivot_balance)
customer_dict.keys()
```

```python
df_test = df_test.set_index('CustomerId') ## не забывать про join
df_test
```

Функция для обучения с разным количеством кластеров

```python
f1_score_models = {}
for key in customer_dict.keys():
  f1 = grid_study(df_test.join(customer_dict[key],how='inner').fillna(0))
  f1_score_models[key] = f1
```
