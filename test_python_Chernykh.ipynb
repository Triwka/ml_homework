## Задание №1

Вам дан текст-отрывок из стихотворения "Дом, который построил Джек", задача написать 2 функции, в первой функции вы будете присваивать номер каждой строчке стихотворения, во 2-ой функции вы должны соединить результат работы 1-ой функции со строкой отрывка. Пример:

* Вот два петуха,
* Которые будят того пастуха,

Результатом 1-ой функции для 2 строчек выше должна быть строки:
* "Это 1 строча в стихотворении про дом Джека - "
* "Это 2 строча в стихотворении про дом Джека - "

Во 2-ой функции нужно обработать результат работы первой функции таким образом:
* "Это 1 строча в стихотворении про дом Джека - Вот два петуха,"
* "Это 2 строча в стихотворении про дом Джека - Которые будят того пастуха,"

Результатом работы 2 функции дожен стать обработанный список, в котором соответствующие приписки с номером строк будут у каждой строчки отрывка.

Отрывок ниже.


```python
otruvok = ['Вот два петуха',
'Которые будят того пастуха',
'Который бранится с коровницей строгою',
'Которая доит корову безрогую',
'Лягнувшую старого пса без хвоста',
'Который за шиворот треплет кота',
'Который пугает и ловит синицу',
'Которая часто ворует пшеницу',
'Которая в тёмном чулане хранится',
'В доме',
'Который построил Джек.']
```

```python
def first_func(list_of_words: list) -> str:
  string = ''
  for i in range(1, len(list_of_words) + 1):
    string = string + (f'Это {i} строчка в стихотворении про дом Джека -\n')
  return string
```

```python
first_func(otruvok)
```

```python
print(first_func(otruvok))
```

```python
def second_func(list_of_words: list) -> list:
  num_list_of_words = []
  for i in range(1, len(list_of_words) + 1):
    num_list_of_words.append(first_func(list_of_words).split('\n')[i - 1] + ' ' + list_of_words[i - 1])
  return num_list_of_words
```

```python
second_func(otruvok)
```

Необходимо в отрывке найти самую длинную строчку и создать матрицу, недостающие элементы заполнить нулями. Пример:
* 1-ая строчка (длинная): Которая в тёмном чулане хранится;
* 2-ая строчка (короче): Который построил Джек.

В длинной строчке по порядку нумеруете все слова - т.е. [1, 2, 3, 4, 5] - получаем 5 слов в самой длинной строке (нумерация строго с единицы).
Короткую строчку преобразуем к виду [1, 2, 3, 0, 0] - т.е. в короткой строчке на 2 слова меньше, чем в длинной, и недостающие нужно добить нулями.

Результатом решения должна стать матрица (список списков), где каждый список будет описывать свою строчку. 

Текст для обработки ниже.

! ОБЯЗАТЕЛЬНОЕ УСЛОВИЕ - решение должно быть универсальным, т.е. отрабатывать аналогично для любого переданного списка строк

```python
test_second_task = ['А это ленивый и толстый пастух',
'Который бранится с коровницей строгою',
'Которая доит корову безрогую',
'Лягнувшую старого пса без хвоста',
'Который за шиворот треплет кота',
'Который пугает и ловит синицу',
'Которая часто ворует пшеницу',
'Которая в тёмном чулане хранится',
'В доме',
'Который построил Джек.']
```

```python
def max_string_len(list_of_words):
  max_len = 0
  for el in list_of_words: #итерируемся по строкам
    len_of_string = len(el.split(' ')) #вычисляем длину каждой строки 
    if max_len <= len_of_string: 
      max_len = len_of_string  #если длина предыдущей строки меньше нынешней, присваиваем значение длины нынешней строки
  return max_len
```

```python
def matrics_list(list_of_words):
  max_len = max_string_len(list_of_words)  #создаем переменную, в которой хранится длина самой большой(по элементам) строки
  matrics_of_numbers = []  #создаем пустой массив, в который будем добавлять элементы
  for el in list_of_words:
    el_of_matrics = [j*0 for j in range(max_len)]  #создаем список размером максимальной строки и заполняем его нулями (можно через np.zeros, но задание вроде как конкретно по кору питона)
    for k in range(1, len(el.split(' ')) + 1):
      el_of_matrics[k - 1] = k   #изменяем элементы соответственно позиции слова в строке
    matrics_of_numbers.append(el_of_matrics)   #добавляем полученный список в матрицу
  return matrics_of_numbers
```

```python
max_string_len(test_second_task)
```

```python
matrics_of_numbers = matrics_list(test_second_task)
matrics_of_numbers
```
