```python
stopwords_rus = stopwords.words("russian")
stemmer = Mystem()
pymorphy = MorphAnalyzer()
#st_words_plus = ['–æ–∫','—à–æ','—ç–º','–±–ª—è',"—ç—Ç–æ","–∑—ã",'—Ñ–∏–≥–∞','–∏–º—Ö–æ',"–≥–æ","–æ—É","—É—É—É—Ö","–æ–º","–≥–≥–≥–≥–º–º–¥–¥","–æ–æ–æ","—Ç–¥","–Ω—É—É","–æ–π","—Ñ—Å","—Ö–∑","—Ç–≥","–º–¥–∞","—á—ë—Ç","—á–µ—Ç","—Ç","–µ","–≥–±","–∞–≥–∞","—ã","–∏–æ","–∞–ø–∏","—â–∞","—á–µ—Ç","–¥—ã–∫",
                 #"—á–µ","–æ—Ö","—Ö–º","–±–ª—è—Ç—å","–º–º–æ","–π–æ","–∑–±—Å","–±","–∫–µ–∫","—É–≥—É","—á–æ—Ç","–±–ª—è–¥—å","—Ç–∏–ø–∞"]
#stopwords_rus.extend(st_words_plus)

def clean_text(text):
    text = re.sub(r"http\S+", "",text) #1)https-–∫–∏

    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF" 
                           "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)  # 2)—Å–º–∞–π–ª–∏–∫–∏
    
    text = re.sub("[A-Za-z]","",text) # 3)–∞–Ω–≥–ª —Å–∏–º–≤–æ–ª—ã

    text = re.sub('\W+',' ', text ) #—É–¥–∞–ª–∏—Ç—å —Å–∏–º–≤–æ–ª—ã

    text = re.sub(r'[‚Ññ"\'_/.:?!\d()%<>;,+#$&\s+]', ' ', text) #–¥–æ–±–∏—Ç—å —É–¥–∞–ª–µ–Ω–∏–µ —Å–∏–º–≤–æ–ª–æ–≤

    text = re.sub(" +", " ", text) #—É–±—Ä–∞—Ç—å –ª–∏—à–Ω–∏–µ –ø—Ä–æ–±–µ–ª—ã

    return text.strip()

def lemmatize_with_pymorphy(tokens):
    lemms = [pymorphy.parse(token)[0].normal_form for token in tokens]  #pymorphy
    return lemms

def del_stopwords(string):
    return ' '.join([word for word in (re.sub(r'[()\s+]', u' ', string)).split() if word not in stopwords_rus])  #stop —Å–ª–æ–≤–∞

def lemmatize(string):
    return ' '.join(stemmer.lemmatize(string)[::2])   #pymystem
```
