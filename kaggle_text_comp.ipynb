```python
stopwords_rus = stopwords.words("russian")
stemmer = Mystem()
pymorphy = MorphAnalyzer()
#st_words_plus = ['ок','шо','эм','бля',"это","зы",'фига','имхо',"го","оу","ууух","ом","ггггммдд","ооо","тд","нуу","ой","фс","хз","тг","мда","чёт","чет","т","е","гб","ага","ы","ио","апи","ща","чет","дык",
                 #"че","ох","хм","блять","ммо","йо","збс","б","кек","угу","чот","блядь","типа"]
#stopwords_rus.extend(st_words_plus)

def clean_text(text):
    text = re.sub(r"http\S+", "",text) #1)https-ки

    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF" 
                           "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)  # 2)смайлики
    
    text = re.sub("[A-Za-z]","",text) # 3)англ символы

    text = re.sub('\W+',' ', text ) #удалить символы

    text = re.sub(r'[№"\'_/.:?!\d()%<>;,+#$&\s+]', ' ', text) #добить удаление символов

    text = re.sub(" +", " ", text) #убрать лишние пробелы

    return text.strip()

def lemmatize_with_pymorphy(tokens):
    lemms = [pymorphy.parse(token)[0].normal_form for token in tokens]  #pymorphy
    return lemms

def del_stopwords(string):
    return ' '.join([word for word in (re.sub(r'[()\s+]', u' ', string)).split() if word not in stopwords_rus])  #stop слова

def lemmatize(string):
    return ' '.join(stemmer.lemmatize(string)[::2])   #pymystem
```

### Эксперимент 1: Pymorphy, без новых стоп слов

```python
train_for_exp_1 = train.copy(deep=True)
test_for_exp_1 = test.copy(deep=True)
```

```python
train_for_exp_1.text = train.text.apply(lambda x: del_stopwords(clean_text(x.lower()))).apply(lambda x: ' '.join(lemmatize_with_pymorphy(x.split())))
test_for_exp_1.text = test.text.apply(lambda x: del_stopwords(clean_text(x.lower()))).apply(lambda x: ' '.join(lemmatize_with_pymorphy(x.split())))

requests.get('https://api.callmebot.com/text.php?user=@mmttry&text=DONE') ## вот эту строчку
```

```python
from sklearn.model_selection import train_test_split
train_descript = train_for_exp_1['text']
test_descript = test_for_exp_1['text']
train_category = train_for_exp_1['label']
X_train, X_test, y_train, y_test = train_test_split(train_descript, train_category, test_size=0.15, random_state=SEED)
%matplotlib inline

def evaluate_vectorizer(vectorizer, train_descript, test_descript, train_category, test_category = None):
    train_vectors = vectorizer.fit_transform(train_descript)
    test_vectors = vectorizer.transform(test_descript)
    
    clf = LinearSVC(random_state=SEED)
    
    clf.fit(train_vectors, train_category)
    
    predictions = clf.predict(test_vectors)
    if (test_category is not None):
        print(classification_report(test_category, predictions))
    
    return predictions

def tokenize_with_razdel(text):
    tokens = [token.text for token in razdel.tokenize(text)]
    
    return tokens

tfidf_vectorizer = TfidfVectorizer(min_df=2, tokenizer=lambda text: tokenize_with_razdel(text))
predictions = evaluate_vectorizer(tfidf_vectorizer, train_descript, test_descript, train_category)
test_predictions = evaluate_vectorizer(tfidf_vectorizer, X_train, X_test, y_train, y_test)
```

```python
test['label'] = predictions
test[['id', 'label']].to_csv('sub_chernykh_svc_pymorphy.csv', index=None) #0.81
```

```python
def clean_text_cat(text):
    #text = re.sub(r"http\S+", "",text) #1)https-ки

    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
        u"\U0001F914"
                               
                           "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)  # 2)смайлики
    
    #text = re.sub("[A-Za-z]","",text) # 3)англ символы

    text = re.sub('\W+',' ', text ) #удалить символы

    text = re.sub(r'[№"\'_/.:?!\d()%<>;,+#$&\s+]', ' ', text) #добить удаление символов

    text = re.sub(" +", " ", text) #убрать лишние пробелы

    return text.strip()

def del_stopwords_cat(string):
    return ' '.join([word for word in (re.sub(r'[()\s+]', u' ', string)).split() if word not in stop_words])  #stop слова
```

```python
train_cat = pd.read_csv('train.csv',sep=';')
test_cat = pd.read_csv('test.csv',sep=';')

train_cat = train_cat.drop(['id'],axis = 1)
test_cat = test_cat.drop(['id'],axis = 1)
```

```python
train_cat.text = train_cat.text.apply(lambda x: del_stopwords_cat(clean_text_cat(x.lower())))
test_cat.text = test_cat.text.apply(lambda x: del_stopwords_cat(clean_text_cat(x.lower())))
```

```python
indexes = []
s = ''
for i in range(0,train_cat.shape[0]):
    s = s + train_cat.text.iloc[i] + ' <brr> ' #тег для сплита
    indexes.append(len(s))

lematized_s = lemmatize(s)  #лематизируем огромную строку

result_s = []
for i in lematized_s.split('brr'): #сплитим по тегу
    result_s.append(i.strip())

train_cat.text = result_s[0:len(result_s)-1]

indexes = []
s = ''
for i in range(0,test_cat.shape[0]):
    s = s + test_cat.text.iloc[i] + ' <brr> '
    indexes.append(len(s))

lematized_s = lemmatize(s)

result_s = []
for i in lematized_s.split('brr'):
    result_s.append(i.strip())

test_cat.text = result_s[0:len(result_s)-1]
```
