## Избавление от дисбаланса классов по методу Дарьи

```python
churn_df = churn_df.dropna().reset_index(drop=True)
X = churn_df.drop('Exited', axis=1) ## датафрейм без несбалансированного таргета
y = churn_df.iloc[:, -1] ## в у хранится несбалансированный таргет
oversample = SMOTE() ## создание экземпляра SMOTE, через который будут использоваться функции класса SMOTE
X_sm, y_sm = oversample.fit_resample(X,y) ## создание новых датафреймов с сбалансированными классами
```

```python
df = X_sm.merge(y_sm, left_index = True, right_index = True) ## приклеивание целевого признака к нецелевым
df
```

```python
df.Exited.value_counts()
```

## Работа с датафреймом Дарьи

```python
df_Darya = df.copy(deep=True)
```

```python
pivot_balance = pd.pivot_table(df_Darya, index = ['CustomerId'],  values = 'Balance')
pivot_balance
```

```python
def dict_df(cluster,df):
  dict_of_df = {}
  df = df.set_index('CustomerId')
  for num in cluster:
    k_means = KMeans(init='k-means++',n_clusters=num)
    k_means.fit(df)
    labels = k_means.labels_
    dict_of_df[str(num) + '_' + 'cluster' + '_' + 'df'] = df.merge(pd.DataFrame(labels,index = df.index).rename(columns = {0:'label'}), left_index = True, right_index = True)
  return dict_of_df
```

```python
test_4cl_Darya = dict_df([2,3,4,5],df)['4_cluster_df']
test_4cl_Darya
```

```python
test_4cl_Darya.label.value_counts()
```

## Работа с датафреймом Егора

```python
df_Egor = df.copy(deep=True)
```

```python
pivot_balance = df_Egor.pivot(columns = 'CustomerId', values=['EstimatedSalary']).droplevel(0,axis=1).fillna(0)
pivot_balance.head()
```

```python
def auto_kmeans(cluster_lst,df):

  clusters_dict = {}
  for cluster in cluster_lst:

    model = KMeans(n_clusters = cluster, random_state = 42)

    clusters = pd.DataFrame([])

    clusters['CustomerId'] = df.columns

    clusters['n'] = model.fit_predict(df.values.T)

    clusters_dict[str(cluster) + '_' + 'cluster' + '_' + 'df'] = clusters.set_index('CustomerId')

  return clusters_dict
```

```python
customer_dict = auto_kmeans([2,3,4,5],pivot_balance)
```

```python
test_4cl_Egor = customer_dict['4_cluster_df']
test_4cl_Egor.n.value_counts()
```
